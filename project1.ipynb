Question 1

```python
'''Develop a Log File Analyzer for simulated system logs.
The program should parse text-based log entries, categorize events, count error types,
analyze user activity, and filter logs based on time or severity.
String splitting, pattern matching, and dictionary-based aggregation must be used. The
application should present analytical summaries via a menu-driven interface.
```

```python
# Our raw data (simulating a file)
logs = [
    "2026-01-06 INFO user_shinoj Login_successful",
    "2026-01-06 ERROR user_biju Disk_space_low",
    "2026-01-06 WARNING user_biju High_memory_usage",
    "2026-01-06 ERROR user_shinoj Unauthorized_access",
    "2026-01-06 ERROR user_thowfeek Connection_timeout"
]
def run_analyzer():
    while True:
        print("\n--- SIMPLE LOG ANALYZER ---")
        print("1. Show Errors  2. Count Users  3. Filter Severity  4. Exit")
        choice = input("Select (1-4): ")

        if choice == "1":
            print("\nList of Errors:")
            for line in logs:
                if "ERROR" in line:
                    print(line)

        elif choice == "2":
            user_counts = {}
            for line in logs:
                user = line.split()[2]
                user_counts[user] = user_counts.get(user, 0) + 1
            print("\nUser Activity:", user_counts)

        elif choice == "3":
            level = input("Enter level (INFO/ERROR/WARNING): ").upper()
            for line in logs:
                if level in line:
                    print(line)

        elif choice == "4":
            print("Goodbye!")
            break

run_analyzer()
```
